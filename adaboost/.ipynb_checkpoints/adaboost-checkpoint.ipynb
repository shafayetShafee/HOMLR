```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_circles
```

```python
np.random.seed(42)
X, y = make_circles(n_samples=500, factor=0.1, noise=0.35, random_state=42)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import cross_val_score

adc = AdaBoostClassifier()

np.mean(cross_val_score(adc, X, y, cv=10, scoring='accuracy'))
```

```python
adc1 = AdaBoostClassifier(n_estimators=100, learning_rate=0.1)
adc1.fit(X, y)

plot_decision_regions(X, y, clf=adc1, legend=1)
```

```python
np.mean(cross_val_score(adc1, X, y, cv=10, scoring='accuracy'))
```

```python
from sklearn.model_selection import GridSearchCV

grid = dict()
grid['n_estimators'] = [10, 50, 100, 500]
grid['learning_rate'] = [0.0001, 0.001, 0.01, 0.1]
grid['algorithm'] = ['SAMME', 'SAMME.R']

grid_search = GridSearchCV(estimator=AdaBoostClassifier(), param_grid=grid, n_jobs=-1, cv=10, scoring='accuracy')

grid_result = grid_search.fit(X, y)

print(f"Best: {grid_result.best_score_} using {grid_result.best_params_}")
```
